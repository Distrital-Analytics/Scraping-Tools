```python
from twitterscraper import query_tweets as q_t
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as po
import plotly.graph_objs as go
import math
%matplotlib inline
from scipy.stats import kurtosis
from scipy.stats import skew
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
bd=dt.date(2020,2,10)
ed=dt.date(2020,2,11)
limit=1000
lang= 'Spanish'
#user=petrogustavo
tw=q_t("ElGanadorEs",begindate=bd, enddate=ed, limit=limit, lang=lang)
```

```python
df=pd.DataFrame(t.__dict__ for t in tw)
df
```

```python
df.head(100)
df.to_csv("datos")
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['time'] = df['timestamp'].dt.time
```

```python
hours = []
cuenta = []
data = pd.DataFrame(columns=["horas","tweets"])
horas = df['time'].to_frame()

for i in range(1,len(horas)):
    hours.append(horas["time"][i].hour)
    
hours.sort()
temp = list(set(hours))

for i in range(0,len(temp)):
    cuenta.append(hours.count(temp[i]))
    temp[i] = temp[i]-5

data["horas"] = temp
data["tweets"] = cuenta

data
```

```python
#si no aparece en jupyter lab, abrir jupyter notebook
data.iplot(x = "horas",y = "tweets")
```
